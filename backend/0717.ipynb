```python
prices:dict[str, float] = {"蘋果": 35.0, "香蕉": 15.5}
```

```python
def process_items(items:list[str]) -> None:
    for item in items:
        print(item.upper())
```

```python
def print_prices(prices:dict[str, float]) -> None:
    for item_name, item_price in prices.items():
        print(f"商品:{item_name}, 價格:{item_price}")
```

**複合與可選型別提示(Union & Optioanal)**

```python
def process_item(item: int | str) -> None:
    print(f"處理資料:{item}")

from typing import Union
def process_item1(item:Union[int, str]) -> None:
    print(item)
```

```python
def say_hi(name: str | None = None) -> None:
    if name is None:
        print("哈囉，世界")
    else:
        print(f"Hello! {name}")
say_hi()
say_hi("Kira")
```

```python
from typing import Optional
def say_hi1(name: Optional[str] = None) -> None:
    if name is None:
        print("哈囉，世界")
    else:
        print(f"Hello! {name}")
say_hi1()
say_hi1("Kira")
```

#Class

```python
#定義一個叫做Person的class
class Person(object):
    # 定義函式
    def __init__(self, name:str):
        self.name = name

john: Person = Person(name="john")

def get_person_name(one_person: Person) -> str:
    return one_person.name

get_person_name(john)
```
