```python
import requests
from requests import Response,JSONDecodeError,HTTPError
from pprint import pprint

url = 'https://data.ntpc.gov.tw/api/datasets/1b72abe8-8862-4130-aeb8-178c1240e6c4/json?page=0&size=10'

try:
    r:Response = requests.request(method='GET',url=url)
    data:list[dict] = r.json()

except JSONDecodeError as e:
    print(f'JSON發生錯誤{e}')
except HTTPError as e:
    print(f'Webapi錯誤{e}')
except Exception as e:
    print(f'發生錯誤:{e}')
```

```python
pprint(data)
```

```python
from pydantic import BaseModel,TypeAdapter

class CameraPosition(BaseModel):
    year:int
    bureau:str
    unit:str
    location:str
adapter = TypeAdapter(list[CameraPosition])
#驗證
list_position = adapter.validate_python(data)
print(repr(list_position))
```

```python
output_jason_str:str = adapter.dump_json(list_position).decode()
print(output_jason_str)
```
